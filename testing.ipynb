```python
import bs4 as bs
import requests
import pandas as pd
```

```python
url = 'https://wish.wis.ntu.edu.sg/webexe/owa/AUS_SCHEDULE.main_display1'

request_params = {
    'staff_access': 'false',
    'acadsem': '2023;2',
    'r_search_type': 'F',
    'boption': 'Search',
    'r_subj_code': ''
    }
```

```python
source = requests.get(url, request_params)
```

```python
soup = bs.BeautifulSoup(source.text, 'lxml')
```

```python
course_df = pd.read_html(soup.prettify())[1]
course_df.head()
```
